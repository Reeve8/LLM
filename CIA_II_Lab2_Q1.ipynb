```python
import numpy as np
import matplotlib.pyplot as plt

def get_positional_encoding(seq_len, d_model):
    """
    Generates the positional encoding matrix.

    Args:
    seq_len (int): Length of the sequence.
    d_model (int): Dimension of the model.

    Returns:
    np.array: Positional encoding matrix of shape (seq_len, d_model)
    """
    # Create a column vector of positions (0, 1, ..., seq_len-1)
    position = np.arange(seq_len)[:, np.newaxis]

    # Calculate the scaling factor for the sine and cosine functions
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))

    # Initialize the positional encoding matrix with zeros
    pe = np.zeros((seq_len, d_model))

    # Apply the sine function to even indices in the matrix
    pe[:, 0::2] = np.sin(position * div_term)

    # Apply the cosine function to odd indices in the matrix
    pe[:, 1::2] = np.cos(position * div_term)

    return pe

def plot_positional_encoding(pe, d_model):
    """
    Plots the positional encoding matrix.

    Args:
    pe (np.array): Positional encoding matrix.
    d_model (int): Dimension of the model.
    """
    plt.figure(figsize=(10, 8))

    # Plot the positional encoding matrix with a color map
    plt.pcolormesh(pe, cmap='viridis')

    # Set the labels and title of the plot
    plt.xlabel('Embedding Dimensions')
    plt.ylabel('Sequence Length')
    plt.title(f'Positional Encoding (d_model = {d_model})')

    # Add a color bar to the plot
    plt.colorbar()
    plt.show()

# Parameters
seq_len = 100  # Length of the sequence

# Dimensionalities to visualize
dimensionalities = [1, 128, 512]

# Generate and plot positional encodings for each dimensionality
for d_model in dimensionalities:
    pe = get_positional_encoding(seq_len, d_model)
    plot_positional_encoding(pe, d_model)
```

this is the visualization of positional encoding, which are used in transformers model to provide information about the positions of tokens in a sequence. The function get_positional_encoding generates a matrix of shape (seq_len, d_model), where each row corresponds to a position in the seq and each column corresponds to a dimension of the model. positional encoding is calculated using sine and cosine functions. the plot  of positional encoding visualise these encodings as colour maps for different dimensionalities, and it provides information is embedded in the models.
